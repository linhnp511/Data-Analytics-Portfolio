Compare performance of 2 segments (AB test)

```python
# Filter relevant records in `df_cashier`
df_cashier_copy = df_cashier_copy[
    (df_cashier_copy['source_type'] == 'ZPI')
    &
    (~df_cashier_copy['domain'].isin(['Funding Channel','Withdraw','Transfer','IBFT','TKTL']))]
```

```python
res = pd.DataFrame(
    columns = [
        'name',
        'old UI',
        'new UI',
    ]
)

res['name'] = ['load_count','submit_count','success_count','submission_rate','interaction_rate','e2e_success_rate']

for index, column in enumerate(['old UI','new UI']):
    if index == 0:
        df = df_cashier_copy[df_cashier_copy['user_id'].apply(lambda x: int(x)%100 in range(80,100))]
    else:
        df = df_cashier_copy[df_cashier_copy['user_id'].apply(lambda x: int(x)%100 in range(30,50))]
    
    load_count = len(df)
    submit_count = len(df[df['is_submitted']==1])
    success_count = len(df[df['is_success']==1])
    submission_rate = submit_count/load_count
    interaction_rate = success_count/submit_count
    e2e_success_rate = submission_rate*interaction_rate

    res[column] = [
        load_count,
        submit_count,
        success_count,
        submission_rate,
        interaction_rate,
        e2e_success_rate
    ]
        
```
